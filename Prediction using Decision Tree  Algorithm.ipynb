```python
import pandas as pd
```

```python
data=pd.read_csv("Iris.csv")
data
```

```python
data=data.drop("Id",axis=1)
data
```

```python
flower_mapping={"Iris-setosa":0,
                "Iris-versicolor":1,
                "Iris-virginica":2
                
    
}
data["Species"]=data["Species"].map(flower_mapping)

features=data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].values
labels=data[['Species']].values
#print("Features: " ,features)
#print("Labels: ",labels)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(features,labels)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,17))
dataviz=tree.plot_tree(dtc,feature_names=features,filled=True,fontsize=20)
```
